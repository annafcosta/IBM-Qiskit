```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
%matplotlib inline

# Com o qc, se define um circuito quntico, neste caso, com 2 qubits e 2 bits clssicos:
qc = QuantumCircuit(2, 2)

# Porta Hadamard
qc.h(0)
qc.h(1)

# Obter o vetor de estado a partir do circuito
state = Statevector.from_instruction(qc)

# Print da esfera de Bloch com o vetor de estado
plot_bloch_multivector(state)
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
%matplotlib inline

# Com o qc, se define um circuito quntico, neste caso, com 2 qubits e 2 bits clssicos:
qc = QuantumCircuit(2, 2)

# Porta Hadamard
qc.h(0)
qc.h(1)

# Medir os qubits para os bits clssicos
qc.measure([0,1],[0,1])

# Print do circuito
qc.draw()
```

```python
from qiskit_ibm_provider import IBMProvider
from qiskit import transpile
from qiskit.visualization import plot_histogram
from qiskit import QuantumCircuit

# Inicializar IBM
provider = IBMProvider() 
backend = provider.get_backend('ibmq_qasm_simulator')

qc = QuantumCircuit(2, 2)
qc.h(0)
qc.h(1)
qc.measure([0, 1], [0, 1]) 

# Transpile o circuito para o backend especificado
transpiled_circuit = transpile(qc, backend)

# Execute o circuito transpilar no backend com 1000 shots
job = backend.run(transpiled_circuit, shots=1000)


result = job.result()
counts = result.get_counts(transpiled_circuit)

# Print das contagens
print(counts)
```

```python
from qiskit_ibm_provider import IBMProvider
from qiskit import transpile
from qiskit.visualization import plot_histogram
from qiskit import QuantumCircuit
import matplotlib.pyplot as plt

provider = IBMProvider() 
backend = provider.get_backend('ibmq_qasm_simulator')

qc = QuantumCircuit(2, 2)
qc.h(0)
qc.h(1)
qc.measure([0, 1], [0, 1])

transpiled_circuit = transpile(qc, backend)

job = backend.run(transpiled_circuit, shots=1000)

result = job.result()

counts = result.get_counts(transpiled_circuit)

plot_histogram(counts)
```
