```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
%matplotlib inline

# Com o qc, se define um circuito quântico, neste caso, com 1 qubit:
qc = QuantumCircuit(1)

# Porta Hadamard
qc.h(0)

# Obter o vetor de estado a partir do circuito
state = Statevector.from_instruction(qc)

# Print da esfera de Bloch com o vetor de estado
plot_bloch_multivector(state)
```

```python
# Resetar o circuito do qubit no estado 1 e depois aplicar uma porta de Hadamard:

qc = QuantumCircuit(1)

qc.x (0)

# Porta Hadamard
qc.h(0)

# Printar o desenho do circuito
qc.draw()
```

```python
# Inicializar o circuito:

qc = QuantumCircuit(1,1)

# Adicionar a porta Hadamard 
qc.h(0)
```

```python
# Definir um circuito de medição com 1 qubit e 1 bit
measurement_circuit = QuantumCircuit(1, 1)

# Função de medição usada para mapear o qubit e o bit pelo seu
# índice no circuito, respectivamente
measurement_circuit.measure(0, 0)

# Concatene os circuitos juntos usando compose()
full_circuit = qc.compose(measurement_circuit)

# Desenhe o circuito completo
full_circuit.draw()
```

```python
from qiskit_ibm_provider import IBMProvider
from qiskit import transpile

# Inicialize o provedor IBM
provider = IBMProvider()  # Inicializa o provedor IBM

# Escolha o backend
backend = provider.get_backend('ibmq_qasm_simulator')

# Transpile o circuito para o backend especificado
transpiled_circuit = transpile(full_circuit, backend)

# Executar o circuito no backend com 1000 shots
job = backend.run(transpiled_circuit, shots=1000)
result = job.result()  # Obter o resultado do trabalho

# Obter as contagens dos resultados do circuito
counts = result.get_counts(transpiled_circuit)

print(counts)
```
